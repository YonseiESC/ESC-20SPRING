### ANOVA F-value 이용

```python
X = wine[wine.columns.difference(['name'])]
y = wine['name']

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
fvalue_selector = SelectKBest(f_classif, k=7)  #feature를 반으로 줄이자 14->7
X_kbest = fvalue_selector.fit_transform(X, y)

print('Original number of features:', X.shape[1])
print('Reduced number of features:', X_kbest.shape[1])
```

```python
X = pd.DataFrame(X_kbest)
X.head() # alchol, colorIntensity, flavanoids, hue, od280_od315, proline, totalPhenols
```

```python
X. columns = ['alchol', 'colorIntensity', 'flavanoids', 'hue', 'od280_od315', 'proline', 'totalPhenols']
X.head()
```

## Train/Test ser split

```python


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=123) #패키지 사용해서 train,test set 나눠주기

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## LDA/QDA

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA().fit(X_train, y_train)
print('LDA score:', lda.score(X_test, y_test))

# QDA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
qda = QDA().fit(X_train,y_train)
print('QDA score:', qda.score(X_test, y_test))
```

```python
np.vstack([y_test, lda.predict(X_test)])
```

## Logistic Regresssion

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='sag', max_iter=10000, multi_class='auto').fit(X_train, y_train) 
#max_iter: 최대 iteration 수 -> 1000을 넣었더니 converge 안됨
print('Logistic Regression score:', clf.score(X_test, y_test))
```

## Logistic Regression with Feature extraction

```python
from sklearn.preprocessing import PolynomialFeatures

from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score  #kcv로 차수 결정위해

for i in range(1,6):
    poly = PolynomialFeatures(degree=i, include_bias=False)
    X_poly = poly.fit_transform(X_train)
    clf2 = LogisticRegression(solver='sag', max_iter=10000, multi_class='auto') #cv=3 이 default
    kcv_score = cross_val_score(clf2, X_poly, y_train) 
    print('%d-degree score:' %i , kcv_score.mean())
```

## Classifier Comparision(score)

```python
poly = PolynomialFeatures(degree=1, include_bias=False)
X_poly = poly.fit_transform(X_train)
clf2 = LogisticRegression(solver='sag', max_iter=10000, multi_class='auto')
kcv_score = cross_val_score(clf2, X_poly, y_train) 

print('LDA score:', lda.score(X_test, y_test))
print('QDA score:', qda.score(X_test, y_test))
print('Logistic Regression score:', clf.score(X_test, y_test))
print('Logi with Feature Extraction score:',kcv_score.mean())

```
