회귀분석 시간에 배운 Y outlier detection 방식 세가지이다.

### Methods of Detecting Outlying Y observations

1) Semi- studentized residuals $e_i* = \frac{e_i}{\sqrt(MSE)}$


2) Internally studentized residuals  $r_i* = \frac{e_i}{s_(e_i)} = \frac{e_i}{\sqrt MSE(1-h_{ii})}$

3) studentized Deleted Residuals $t_i  = \frac{d_i}{s_(d_i)} = e_i [\frac{n-p-1}{SSE(1-h_{ii})- e_i^2}]^{\frac{1}{2}} \sim t(n-p-1)$

### Cramer's V 이용

```python
import scipy.stats as ss
import itertools
```

```python
# Compute the correlation matrix
def cramers_corrected_stat(confusion_matrix):
    """ calculate Cramers V statistic for categorical-categorical association.
        uses correction from Bergsma and Wicher, 
        Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))

nominal_list=['HTYP','HLIV','REL','RLI','Q1','Q2B','Q3A1','Q14A1', 'Q14A2','Q17A1','Q17A2','Q18','Q19A1','Q19A2','Q21A','Q21B1','Q21B2','Q24','Q25A1','Q25A2','Q26','Q31A1','Q31A2','Q38','Q39','Q42','Q44','Q47C','DE1','DE2','DE3','DE5','DE6','DE7','DE9']
happy_nominal=happy[nominal_list]
corrM = np.zeros((len(nominal_list),len(nominal_list)))
# there's probably a nice pandas way to do this
for col1, col2 in itertools.combinations(nominal_list, 2):
    idx1, idx2 = nominal_list.index(col1), nominal_list.index(col2)
    corrM[idx1, idx2] = cramers_corrected_stat(pd.crosstab(happy_nominal[col1], happy_nominal[col2]))
    corrM[idx2, idx1] = corrM[idx1, idx2]

corr_nominal = pd.DataFrame(corrM, index=nominal_list, columns=nominal_list)
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr_nominal, dtype=np.bool) # Upper triangular로 만드는거라네요
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(30,30))

# Draw the heatmap with the mask and correct aspect ratio
vmax = np.abs(corr_nominal.values[~mask]).max()
sns.heatmap(corr_nominal, mask=mask, cmap=plt.cm.PuOr, vmin=-vmax, vmax=vmax, annot=True,
            square=True, linecolor="lightgray", linewidths=1, ax=ax, cbar_kws={"shrink": .2})

for i in range(len(corr_nominal)): # 위치 조절
    ax.text(i+0.5,(i+0.5), corr_nominal.columns[i], 
            ha="center", va="center", rotation=45)
    for j in range(i+1, len(corr_nominal)):
        s = "{:.3f}".format(corr_nominal.values[i,j])
        ax.text(j+0.5,i+0.5,s, 
            ha="center", va="center")

plt.show()
```
